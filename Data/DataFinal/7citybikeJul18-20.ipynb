# CITY BIKE DATA JULY 2018-2020

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
jul2018 = "7jul/JC-201807-citibike-tripdata.csv"
jul2019 = "7jul/JC-201907-citibike-tripdata.csv"
jul2020 = "7jul/JC-202007-citibike-tripdata.csv"
```

```python
# Read July 2018, 2019 & 2020 Citi Bike Data Files & Store into Pandas DataFrames
jul2018_df = pd.read_csv(jul2018, encoding="ISO-8859-1")
jul2019_df = pd.read_csv(jul2019, encoding="ISO-8859-1")
jul2020_df = pd.read_csv(jul2020, encoding="ISO-8859-1")
```

```python
# Combine All Three July DataFrames from 2018-2020 Into Single DataFrame
combojuldf = jul2018_df.append([jul2019_df, jul2020_df], ignore_index=True)
combojuldf.head()
```

```python
# Rename Columns
renamecombojuldf = combojuldf.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
renamecombojuldf.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
renamecombojuldf["Start Time"] = pd.to_datetime(renamecombojuldf["Start Time"])
renamecombojuldf["Stop Time"] = pd.to_datetime(renamecombojuldf["Stop Time"])
renamecombojuldf.head()
```

```python
# Save Combined DataFrames to CSV
renamecombojuldf.to_csv("7citibike-tripdataJul18-20.csv", index=False, header=True)
```
