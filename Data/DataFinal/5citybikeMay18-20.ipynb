# CITY BIKE DATA MAY 2018-2020

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
may2018 = "5may/JC-201805-citibike-tripdata.csv"
may2019 = "5may/JC-201905-citibike-tripdata.csv"
may2020 = "5may/JC-202005-citibike-tripdata.csv"
```

```python
# Read May 2018, 2019 & 2020 Citi Bike Data Files & Store into Pandas DataFrames
may2018_df = pd.read_csv(may2018, encoding="ISO-8859-1")
may2019_df = pd.read_csv(may2019, encoding="ISO-8859-1")
may2020_df = pd.read_csv(may2020, encoding="ISO-8859-1")
```

```python
# Combine All Three May DataFrames from 2018-2020 Into Single DataFrame
combomaydf = may2018_df.append([may2019_df, may2020_df], ignore_index=True)
combomaydf.head()
```

```python
# Rename Columns
renamecombomaydf = combomaydf.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
renamecombomaydf.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
renamecombomaydf["Start Time"] = pd.to_datetime(renamecombomaydf["Start Time"])
renamecombomaydf["Stop Time"] = pd.to_datetime(renamecombomaydf["Stop Time"])
renamecombomaydf.head()
```

```python
# Save Combined DataFrames to CSV
renamecombomaydf.to_csv("5citibike-tripdataMay18-20.csv", index=False, header=True)
```
