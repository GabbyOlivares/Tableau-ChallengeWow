# CITY BIKE DATA MARCH 2018-2020

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
mar2018 = "3mar/JC-201803-citibike-tripdata.csv"
mar2019 = "3mar/JC-201903-citibike-tripdata.csv"
mar2020 = "3mar/JC-202003-citibike-tripdata.csv"
```

```python
# Read March 2018, 2019 & 2020 Citi Bike Data Files & Store into Pandas DataFrames
mar2018_df = pd.read_csv(mar2018, encoding="ISO-8859-1")
mar2019_df = pd.read_csv(mar2019, encoding="ISO-8859-1")
mar2020_df = pd.read_csv(mar2020, encoding="ISO-8859-1")
```

```python
# Combine All Three March DataFrames from 2018-2020 Into Single DataFrame
combomardf = mar2018_df.append([mar2019_df, mar2020_df], ignore_index=True)
combomardf.head()
```

```python
# Rename Columns
renamecombomardf = combomardf.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
renamecombomardf.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
renamecombomardf["Start Time"] = pd.to_datetime(renamecombomardf["Start Time"])
renamecombomardf["Stop Time"] = pd.to_datetime(renamecombomardf["Stop Time"])
renamecombomardf.head()
```

```python
# Save Combined DataFrames to CSV
renamecombomardf.to_csv("3citibike-tripdataMar18-20.csv", index=False, header=True)
```
