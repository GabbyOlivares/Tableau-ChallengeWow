# CITY BIKE DATA NOVEMBER 2018-2020

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
nov2018 = "11nov/JC-201811-citibike-tripdata.csv"
nov2019 = "11nov/JC-201911-citibike-tripdata.csv"
nov2020 = "11nov/JC-202011-citibike-tripdata.csv"
```

```python
# Read November 2018, 2019 & 2020 Citi Bike Data Files & Store into Pandas DataFrames
nov2018_df = pd.read_csv(nov2018, encoding="ISO-8859-1")
nov2019_df = pd.read_csv(nov2019, encoding="ISO-8859-1")
nov2020_df = pd.read_csv(nov2020, encoding="ISO-8859-1")
```

```python
# Combine All Three November DataFrames from 2018-2020 Into Single DataFrame
combonovdf = nov2018_df.append([nov2019_df, nov2020_df], ignore_index=True)
combonovdf.head()
```

```python
# Rename Columns
renamecombonovdf = combonovdf.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
renamecombonovdf.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
renamecombonovdf["Start Time"] = pd.to_datetime(renamecombonovdf["Start Time"])
renamecombonovdf["Stop Time"] = pd.to_datetime(renamecombonovdf["Stop Time"])
renamecombonovdf.head()
```

```python
# Save Combined DataFrames to CSV
renamecombonovdf.to_csv("11citibike-tripdataNov18-20.csv", index=False, header=True)
```
