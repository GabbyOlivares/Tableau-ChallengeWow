# CITY BIKE DATA AUGUST 2018-2020

```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
aug2018 = "8aug/JC-201808-citibike-tripdata.csv"
aug2019 = "8aug/JC-201908-citibike-tripdata.csv"
aug2020 = "8aug/JC-202008-citibike-tripdata.csv"
```

```python
# Read August 2018, 2019 & 2020 Citi Bike Data Files & Store into Pandas DataFrames
aug2018_df = pd.read_csv(aug2018, encoding="ISO-8859-1")
aug2019_df = pd.read_csv(aug2019, encoding="ISO-8859-1")
aug2020_df = pd.read_csv(aug2020, encoding="ISO-8859-1")
```

```python
# Combine All Three August DataFrames from 2018-2020 Into Single DataFrame
comboaugdf = aug2018_df.append([aug2019_df, aug2020_df], ignore_index=True)
comboaugdf.head()
```

```python
# Rename Columns
renamecomboaugdf = comboaugdf.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
renamecomboaugdf.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
renamecomboaugdf["Start Time"] = pd.to_datetime(renamecomboaugdf["Start Time"])
renamecomboaugdf["Stop Time"] = pd.to_datetime(renamecomboaugdf["Stop Time"])
renamecomboaugdf.head()
```

```python
# Save Combined DataFrames to CSV
renamecomboaugdf.to_csv("8citibike-tripdataAug18-20.csv", index=False, header=True)
```
